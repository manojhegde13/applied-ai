```python
# you should not import any other packages
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings("ignore")
import numpy as np
from sklearn.linear_model import SGDRegressor
```

```python
import numpy as np
import scipy as sp
import scipy.optimize

def angles_in_ellipse(num,a,b):
    assert(num > 0)
    assert(a < b)
    angles = 2 * np.pi * np.arange(num) / num
    
    print(angles)
    if a != b:
        e = (1.0 - a ** 2.0 / b ** 2.0) ** 0.5
        print(e)
        tot_size = sp.special.ellipeinc(2.0 * np.pi, e)
        print(tot_size)
        arc_size = tot_size / num
        arcs = np.arange(num) * arc_size
        print(arcs)
        res = sp.optimize.root(
            lambda x: (sp.special.ellipeinc(x, e) - arcs), angles)
        angles = res.x 
    print(angles)
    return angles
    

a = 2
b = 9
n = 50

phi = angles_in_ellipse(n, a, b)
e = (1.0 - a ** 2.0 / b ** 2.0) ** 0.5
arcs = sp.special.ellipeinc(phi, e)

fig = plt.figure()
ax = fig.gca()
ax.axes.set_aspect('equal')
ax.scatter(b * np.sin(phi), a * np.cos(phi))
plt.show()
```

```python
alpha=[0.001,1,100]
s=0
ratios = [(0,2),(21, 13), (-23, -15), (22,14), (23, 14)] 
plt.figure(figsize=(20,15))

for reg in alpha:
    X_n = X
    Y_n = Y
    
    
    for j,i in enumerate(ratios):
        s = s+1
        X_n=np.append(X_n,i[0]).reshape(-1,1)
        Y_n=np.append(Y_n,i[1]).reshape(-1,1)
        clf=SGDRegressor(alpha=reg, eta0=0.001, learning_rate='constant',random_state=10)
        clf.fit(X_n,Y_n)
        
        predict=clf.predict(X_n)
        
        plt.subplot(3,5,s)

        plt.scatter(X_n,Y_n)
        plt.plot(X_n,predict,color="black")


                    
```
